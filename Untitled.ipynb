## Plotting with matplotlib

```python
#import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define the potential energy
def potential_energy(x, k=1.0):
        return 0.5 * k * x**2 #k=spring constant

#test function for the potential energy
potential_energy(2,10)
```

```python
#Define mesh for plotting
x = np.linspace(-5, 5, 100) #x_min, x_max, number of point

#Define y values as a function of x
y = potential_energy(x)

print(np.transpose([x, y]))
```

```python
#Plot the graph
plt.figure(figsize=(2,2)) #Size of graph in inches
plt.plot(x,y, color="black", linewidth=.5)

#Assign axes values
plt.xlabel("displacement (arbitrary units)")

plt.ylabel("potential energy (arbitrary units)")

# Change limits on graph
plt.xlim([-4,4])
plt.ylim([-1,10])

plt.show() #combine all elements
```

## Contour Plots

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt

def potential_energy_2D(x, y, kx=1.0, ky=1.0):
        return 0.5 * (kx * x**2 + ky * y**2) #k=spring constant

#test function for the potential energy
potential_energy_2D(2,1)
```

```python
#define meshes for plotting
x=np.linspace(-5, 5, 3)
y=np.linspace(-5, 5, 3)

X, Y = np.meshgrid(x,y)

Z=potential_energy_2D(X,Y, 1, 4)

print(x)
print()
print(Y)
print()
print(Z)
```

```python
# Construct contour plot
plt.contourf(X, Y, Z)
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
def U(Q1, Q2, K1, K2, C):
    return 0.5 * K1 * Q1**2 + 0.5 * K2 * Q2**2 + C * Q1 * Q2**2
```

```python
def plot_energy_fixed_Q2(K1, K2, C):
    Q1 = np.linspace(-10, 10, 400)
    Q2_values = [-5, -2, 0, 2, 5]  # Fixed Q2 values

    plt.figure(figsize=(12, 6))

    # First plot: U(Q1, Q2) with varying Q2
    plt.subplot(1, 2, 1)
    U0 = U(Q1, 0, K1, K2, C)
    plt.plot(Q1, U0, label='Q2=0', color='black')
    
    for Q2 in Q2_values:
        U_Q2 = U(Q1, Q2, K1, K2, C)
        plt.plot(Q1, U_Q2, label=f'Q2={Q2}')
    
    plt.title('Potential Energy U(Q1, Q2) with varying Q2')
    plt.xlabel('Q1')
    plt.ylabel('U(Q1, Q2)')
    plt.axhline(0, color='gray', lw=0.5, ls='--')
    plt.legend()
    plt.grid()

    # Second plot: 2D contour plot
    plt.subplot(1, 2, 2)
    Q1_grid, Q2_grid = np.meshgrid(np.linspace(-10, 10, 400), np.linspace(-10, 10, 400))
    U_grid = U(Q1_grid, Q2_grid, K1, K2, C)
    
    contour = plt.contourf(Q1_grid, Q2_grid, U_grid, levels=50, cmap='viridis')
    plt.colorbar(contour)
    plt.title('Contour Plot of U(Q1, Q2)')
    plt.xlabel('Q1')
    plt.ylabel('Q2')
    
    plt.tight_layout()
    plt.show()
```

```python
def interactive_plot(K1=1.0, K2=1.0, C=0.1):
    plot_energy_fixed_Q2(K1, K2, C)

interact(interactive_plot,
         K1=FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='K1'),
         K2=FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='K2'),
         C=FloatSlider(value=0.1, min=-1.0, max=1.0, step=0.01, description='C'));
```

# Answers to Questions

1. As you increase C, how do the low-energy contours start to change?
   The geometry of the contour plot shifts much more drastically as C increases when K1 and K2 are lower. The plot starts with the rightmost corners  having the highest U values, then will suddenly flip around C=.1 to being concentrated in the leftmost corners then steadily increases to higher U values in the leftmost corners as C approaches its maximum.
2. Does the force on Q1 depend on the direction of Q2?
   Yes, the force on Q1 does depend on the direction of Q2.
3. What do you expect the influence of driving Q2 to be on the motion of Q1?
   Depending on the value of C, as Q2 varies, Q1 will either be in sync with Q2 or will be out of phase with Q2.
4. If Q1 is kicked away from equilibrium, what do you expect its influence to be on Q2?
   If Q1 is kicked form equilibrium, it will alter the motion of Q2 by the coupling term CQ1Q2^2. This shows how connected Q1 and Q2 are, and that there is a strong coupling between their motions.
